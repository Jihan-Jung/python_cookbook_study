```python
tot_data = []
cur_sum = 0
for (chap,num) in enumerate(chap_tot):
    for i in range(1,num+1):
        cur_sum += 1
        tot_data.append((chap+1,i,cur_sum))
```

```python
## tot_mem will be 4 or 5

def print_mypart_by_week(data,position,tot_mem,week):
    
    tot_size = 260 # 260개 질문
    
    if position == tot_mem:
        position = 0
        
    mypart = []
    for (chap,num,idx) in data:
        if idx % tot_mem == position:
            mypart.append((chap,num))
            
    if tot_mem == 4:
        week_1_cnt = 9
        other_week_cnt = 8
        
        if week == 1:
            for (c,n) in mypart[:week_1_cnt]:
                print((c,n))
        elif tot_size > tot_mem*(week_1_cnt+(other_week_cnt)*(week-2)):
            st_pt = week_1_cnt + (week-2)*other_week_cnt
            for (c,n) in mypart[st_pt:(st_pt+other_week_cnt)]:
                print((c,n))
        else:
            print("error : out of week")
            
    elif tot_mem == 5:
        week_until_4_cnt = 7
        other_week_cnt = 6
        
        if week <= 4:
            st_pt = (week-1)*week_until_4_cnt
            for (c,n) in mypart[st_pt:(st_pt+week_until_4_cnt)]:
                print((c,n))
                
        elif tot_size > tot_mem*(4*week_until_4_cnt + (other_week_cnt)*(week-5)):
            st_pt = 4*week_until_4_cnt + (week-5)*other_week_cnt
            for (c,n) in mypart[st_pt:(st_pt+other_week_cnt)]:
                print((c,n))                
        else:
            print("error : out of week")
```

```python
print_mypart_by_week(data=tot_data,position=0,tot_mem=4,week=1)
```
